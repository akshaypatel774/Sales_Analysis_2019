```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

### Merging 12 months data

```python
files = [file for file in os.listdir(f'D:/PythonProjects/SalesAnalysis')]
data = pd.DataFrame()
for file in files:
    df = pd.read_csv(f'D:/PythonProjects/SalesAnalysis/{file}')
    data = pd.concat([data, df])

data.head()
```

```python
# Get all nan data
nan_df = data[data.isna().any(axis=1)]
nan_df.head()
```

```python
# Drop all nan data
data = data.dropna()
data.head()
```

```python
# Where Order Date = Order Date
invalid_date = data.loc[data['Order Date'] == 'Order Date']
invalid_date.head()
```

```python
# Removing rows with 'Order Date' as Order Date
data =  data.loc[data['Order Date'] != 'Order Date']
data.head()
```

```python
# Extracting month column from order date column
data['Month'] = data['Order Date'].str[0:2]
data['Month'] = data['Month'].astype('int')
data.head()
```

```python
# Converting columns below to numerical values
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'])
data['Price Each'] = pd.to_numeric(data['Price Each'])
```

```python
# Adding new Sales column
data['Sales'] = data['Quantity Ordered'] * data['Price Each']
data.head()
```

```python
# Total sales during a month
total_sales = data.groupby('Month').sum(numeric_only=True)
total_sales.sort_values(['Sales'], ascending=False)
```

```python
months = range(1,13)
plt.bar(months, total_sales['Sales'])
plt.xticks(months)
plt.title('Sales by Months (2019)')
plt.xlabel('Months')
plt.ylabel('Sales in USD($)')
```

```python
# Adding City Column with state name
data['City'] = data['Purchase Address'].apply(lambda x: f"{x.split(',')[1]}, {x.split(',')[2].split(' ')[1]}")
data.head()
```

```python
# Total sales of each City
total_sales = data.groupby('City').sum(numeric_only=True)
total_sales
```

```python
# Cannot use as it gets sorted differently --cities = data['City'].unique()
cities = [city for city, df in data.groupby('City')]
plt.bar(cities, total_sales['Sales'])
plt.xticks(cities, rotation='vertical', size=7)
plt.title('Sales by Cities (2019)')
plt.xlabel('City Names')
plt.ylabel('Sales in USD($)')
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
# Add Hour column
data['Hour'] = data['Order Date'].dt.hour
```

```python
# Orders made during each hour in 2019
data.groupby(['Hour']).count()['Sales']
```

```python
hours = [hour for hour, df in data.groupby('Hour')]
plt.plot(hours, data.groupby(['Hour']).count(), color='green')
plt.xticks(hours)
plt.title('Orders by Hours')
plt.xlabel('Hours')
plt.ylabel('Total Orders')
plt.grid()

plt.show()
```

### To get all products that are ordered together

```python
df = data[data['Order ID'].duplicated(keep=False)] # keep= False keeps all instances instead of removing first or the last instance.
df.head()
```

```python
# Concatenating products bought together
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2))) # For 2 items

# print(count)
for key, value in count.most_common(5):
    print(key, value)

# iPhone and Lightning Charging Cable are 2 most bought items in pairs (1005 times)
# Google Phone, USB-C Charging Cable, and Wired Headphones are 3 most bought items in pairs (87 times)
```

```python
# Get number of products sold for each product
data.groupby('Product').sum(numeric_only=True)['Quantity Ordered']
```

```python
products_list = data.groupby('Product')
counts = products_list.sum(numeric_only=True)['Quantity Ordered']

products = [product for product, df in products_list]
plt.bar(products, counts)
plt.xticks(products, rotation='vertical', size=7)
plt.title('Products sold in 2019')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.show()

# AAA Batteries are the most sold products, maybe because of low price comparatively
```

```python
prices = data.groupby('Product').mean(numeric_only=True)['Price Each']
prices.head()
```

```python
fig, ax1 =plt.subplots()
ax2 = ax1.twinx()
ax1.bar(products, counts, color='g')
ax2.plot(products, prices, color='r')
plt.title('Products sold in 2019 with Prices')
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Sold', color='g')
ax2.set_ylabel('Price', color='r')
ax1.set_xticklabels(products, rotation='vertical', size=7)

plt.show()
```
